#Eksik Değer Analizi
Veri setindeki eksik değerlerin tespiti ve bu değerlerle ilgili uygulanabilecek yaklaşımlar açıklanır.

```python
df.isnull().sum()
```

```python
#Ortalama ile eksik değerleri doldurma
# 1. Veri çakışmalarını önlemek için orijinal veriyi kopyala
df_clone = df.copy()

# 2. Eksik değerleri ortalama ile doldur
df_clone["Arrival Delay in Minutes"] = df_clone["Arrival Delay in Minutes"].fillna(
    df_clone["Arrival Delay in Minutes"].mean()
)
# 3. Eksik değer kalıp kalmadığını kontrol et
print("Eksik değerin doldurulduğu değer:",df_clone["Arrival Delay in Minutes"].mean())

# 4. Eksik değer kalıp kalmadığını kontrol et
print("Eksik değer sayısı:", df_clone["Arrival Delay in Minutes"].isnull().sum())
```

```python
df_clone["Arrival Delay in Minutes"]
```

```python
#Medyan değeri ile eksik değerleri doldurmak
# 1. Veri çakışmalarını önlemek için orijinal veriyi kopyala
df_clone1 = df.copy()

# 2. Eksik değerleri medyan ile doldur
df_clone1["Arrival Delay in Minutes"] = df_clone1["Arrival Delay in Minutes"].fillna(
    df_clone1["Arrival Delay in Minutes"].median()
)

# 3. Doldurulan değer (medyan)
print("Eksik değerin doldurulduğu değer:", df_clone1["Arrival Delay in Minutes"].median())

# 4. Eksik değer kontrolü
print("Eksik değer sayısı:", df_clone1["Arrival Delay in Minutes"].isnull().sum())
```

```python
df_clone1["Arrival Delay in Minutes"]
```

```python
#Sınıfa göre ortalama değeri ile eksik değerleri doldurmak
# 1. Kopya oluştur
df_clone2 = df.copy()

# 2. Sınıfa göre ortalama ile doldur
df_clone2["Arrival Delay in Minutes"] = df_clone2.groupby("Class")["Arrival Delay in Minutes"].transform(
    lambda x: x.fillna(x.mean())
)

# 3. Ortalama değer gösterimi (tüm sınıflara göre genel ortalama)
print("Genel ortalama:", df_clone2["Arrival Delay in Minutes"].mean())

# 4. Kontrol
print("Eksik değer sayısı:", df_clone2["Arrival Delay in Minutes"].isnull().sum())
```

```python
#KNN değeri ile eksik değerleri doldurmak
from sklearn.impute import KNNImputer

# 1. Kopya oluştur
df_clone3 = df.copy()

# 2. Sayısal verileri ayır (KNN sayısal verilerle çalışır)
numeric_cols = df_clone3.select_dtypes(include=["float64", "int64"]).columns

# 3. KNN imputer uygula
imputer = KNNImputer(n_neighbors=5)
df_clone3[numeric_cols] = imputer.fit_transform(df_clone3[numeric_cols])

# 4. Kontrol
print("Eksik değer sayısı:", df_clone3["Arrival Delay in Minutes"].isnull().sum())
```

#Aykırı Değer (Outlier) Analizi
Sayısal değişkenler üzerinde aykırı değerler belirlenir ve görsel/istatistiksel yöntemlerle raporlanır.


```python
#IQR (Interquartile Range) Yöntemi

df_clone_iqr = df_clone.copy()
numeric_cols = df_clone_iqr.select_dtypes(include=["number"]).columns

# Aykırı değer sayısını tutacak liste
iqr_outlier_counts = []

for col in numeric_cols:
    Q1 = df_clone_iqr[col].quantile(0.25)
    Q3 = df_clone_iqr[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_count = df_clone_iqr[(df_clone_iqr[col] < lower) | (df_clone_iqr[col] > upper)].shape[0]
    iqr_outlier_counts.append((col, outlier_count))

    # Aykırıları veri dışı bırak
    df_clone_iqr = df_clone_iqr[(df_clone_iqr[col] >= lower) & (df_clone_iqr[col] <= upper)]

# Her değişkendeki aykırı sayısını yazdır
print("📊 IQR yöntemiyle aykırı değer sayıları:")
for col, count in iqr_outlier_counts:
    print(f"- {col}: {count} aykırı değer")

print("✅ Temizlenmiş veri satır sayısı (IQR):", df_clone_iqr.shape[0])
```

```python
# Sayısal sütunları seç
numeric_cols = df_clone.select_dtypes(include=["number"]).columns

# Her değişken için IQR yöntemiyle aykırı değerleri scatter plot ile görselleştir
for col in numeric_cols:
    Q1 = df_clone[col].quantile(0.25)
    Q3 = df_clone[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_mask = (df_clone[col] < lower) | (df_clone[col] > upper)

    plt.figure(figsize=(6, 4))
    plt.scatter(df_clone.index[~outlier_mask], df_clone.loc[~outlier_mask, col],
                label="Normal", alpha=0.5)
    plt.scatter(df_clone.index[outlier_mask], df_clone.loc[outlier_mask, col],
                color="red", label="Aykırı", alpha=0.8)
    plt.xlabel("Gözlem (index)")
    plt.ylabel(col)
    plt.title(f"{col} - IQR Aykırı Değer Nokta Grafiği")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
#Z-Score Yöntemi
from scipy.stats import zscore

df_clone_z = df_clone.copy()
numeric_cols = df_clone_z.select_dtypes(include=["number"]).columns

# Z-skorlarını hesapla
z_scores = np.abs(zscore(df_clone_z[numeric_cols]))

# Aykırı değer sayısını her sütun için hesapla
z_outlier_counts = (z_scores > 3).sum(axis=0)

print("📊 Z-Score yöntemiyle aykırı değer sayıları:")
for col, count in zip(numeric_cols, z_outlier_counts):
    print(f"- {col}: {count} aykırı değer")

# Z-score'a göre aykırı satırları kaldır
df_clone_z = df_clone_z[(z_scores < 3).all(axis=1)]
print("✅ Temizlenmiş veri satır sayısı (Z-Score):", df_clone_z.shape[0])
```

```python
# Sayısal sütunları seç
numeric_cols = df_clone.select_dtypes(include=["number"]).columns

# Z-score hesapla
z_scores = np.abs(zscore(df_clone[numeric_cols]))

# Her değişken için scatter plot (nokta grafiği) oluştur
for col in numeric_cols:
    col_z = z_scores[:, numeric_cols.get_loc(col)]
    outlier_mask = col_z > 3

    plt.figure(figsize=(6, 4))
    plt.scatter(x=df_clone.index[~outlier_mask], y=df_clone[~outlier_mask][col],
                label="Normal", alpha=0.5)
    plt.scatter(x=df_clone.index[outlier_mask], y=df_clone[outlier_mask][col],
                color="red", label="Aykırı", alpha=0.8)
    plt.xlabel("Gözlem (index)")
    plt.ylabel(col)
    plt.title(f"{col} - Z-Score Aykırı Değer Nokta Grafiği")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

#Görselleştirme
Sayısal ve kategorik değişkenler için uygun grafik türleri kullanılarak veriler görselleştirilir.

```python
# IQR Yöntemine Göre Temizlenmiş Veride Görselleştirme
# 1. Sayısal değişkenler için histogram + boxplot
print("📊 IQR ile temizlenmiş sayısal değişkenler:")
for col in df_clone_iqr.select_dtypes(include=["number"]).columns:
    plt.figure(figsize=(12, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df_clone_iqr[col], kde=True, color='skyblue')
    plt.title(f"{col} - Histogram")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df_clone_iqr[col], color='lightgreen')
    plt.title(f"{col} - Boxplot")

    plt.tight_layout()
    plt.show()

# 2. Kategorik değişkenler için countplot
print("📊 IQR ile temizlenmiş kategorik değişkenler:")
for col in df_clone_iqr.select_dtypes(include=["object", "category"]).columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df_clone_iqr, x=col, palette="pastel")
    plt.title(f"{col} - Kategori Dağılımı")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 2. Z-Score Yöntemine Göre Temizlenmiş Veride Görselleştirme
# 1. Sayısal değişkenler için histogram + boxplot
print("📊 Z-Score ile temizlenmiş sayısal değişkenler:")
for col in df_clone_z.select_dtypes(include=["number"]).columns:
    plt.figure(figsize=(12, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df_clone_z[col], kde=True, color='salmon')
    plt.title(f"{col} - Histogram")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df_clone_z[col], color='orange')
    plt.title(f"{col} - Boxplot")

    plt.tight_layout()
    plt.show()

# 2. Kategorik değişkenler için countplot
print("📊 Z-Score ile temizlenmiş kategorik değişkenler:")
for col in df_clone_z.select_dtypes(include=["object", "category"]).columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df_clone_z, x=col, palette="Set2")
    plt.title(f"{col} - Kategori Dağılımı")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
